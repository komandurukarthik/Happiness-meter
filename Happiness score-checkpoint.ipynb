Now all the data we got is liner so we will try to find the Relation between all the independent and Dependent variables using visualization techniques.

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1
for column in df:
    if plotnumber<=10:
        ax=plt.subplot(2,5,plotnumber)
        sn.distplot(df[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

As shown above there is skew ness in the standarderror,Generosity,Trust so we need to Remove the Skewness for those Fields

Removing Skewness of above fields

```python
q=df['Generosity'].quantile(0.98)
data_cleaned=df[df['Generosity']<q]
```

Now let us see how the data set looks like after removing the Skewness

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1
for column in data_cleaned:
    if plotnumber<=10:
        ax=plt.subplot(2,5,plotnumber)
        sn.distplot(data_cleaned[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```

As seen above there is no skewness in our data so we can proceed further.

```python
x=df.drop(columns=['Happiness Score'],axis=1)
y=df['Happiness Score']
```

Before building a model let us Try to Visualize the relation ship between the dependent and independent variables

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1
for column in x:
    if plotnumber<=9:
        ax=plt.subplot(2,5,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Happiness Score',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

By above Distribution we were able to find that the relationship between the dependent and independent variable looks linear now lets scale our data for this iam using standard Scalar

```python
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

The x_scaled gives us a numpy array

Now let us go a head and use linear to find out how good it fit our data for this iam splitting my data into train data and test data using train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=21)
```

```python
y_train.head()
```

Finding the best random state use here 

So now Lets instantiate our model and Train it for this iam using liner regression

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

Let us find the R square value for the Trained model

```python
regression.score(x_test,y_test)
```

My modelgave 99.9% accuracy when used adjusted R square technique.

```python
y_pred=regression.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

Both the models are giving 99% data accuracy no applying regularization Techniques
